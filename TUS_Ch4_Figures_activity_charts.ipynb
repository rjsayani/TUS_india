```python
##Maharashtra activity chart

# Get number of rows in times array
times = np.arange(0,24,0.5)
nt = times.shape[0]

# Set minimum and maximum state values and calculate total number of states
state_min = 1
state_max = 10
nstate_tot = state_max - state_min + 1

# Set the sleeping state value
sleeping_state = 10

# Create an empty list to hold activity time matrix
activity_time_matrix_MH = [np.zeros(nt) for j in range(nstate_tot)]

# Get unique IDs from TUS_MH dataframe
ids = np.sort(TUS_MH_rural.index.unique())

# Loop through each ID
for idi in ids:
    # Create a temporary dataframe for the current ID
    TUS_MH_temp0 = TUS_MH_rural[TUS_MH_rural.index == idi].copy()

    # Get unique activity groups for the current ID
    activities = TUS_MH_temp0['Activity group'].unique()

    # Loop through each activity group
    for activity in activities:
        # Create a temporary dataframe for the current activity group
        TUS_MH_temp = TUS_MH_temp0[TUS_MH_temp0['Activity group'] == activity].copy()

        # Get the number of entries in the temporary dataframe
        nentries = TUS_MH_temp.index.nunique()

        # Loop through each entry
        for i in range(nentries):
            # Get start and end times for the current entry
            time_stamp1 = TUS_MH_temp['Start_time'].iloc[i]
            time_stamp2 = TUS_MH_temp['End_time'].iloc[i]

            # Calculate the time positions for start and end times
            timepos1 = np.where(times == (time_stamp1.hour + time_stamp1.minute / 60))[0][0]
            timepos2 = np.where(times == (time_stamp2.hour + time_stamp2.minute / 60))[0][0]

            # Check if end time is earlier than start time
            if time_stamp2.hour < time_stamp1.hour:
                # Loop through time positions from start time to end of times array
                for tp in np.arange(timepos1, nt, 1):
                    activity_time_matrix_MH[activity - 1][tp] += 1 

                # Loop through time positions from beginning of times array to end time
                for tp in np.arange(0, timepos2 + 1, 1):
                    activity_time_matrix_MH[activity - 1][tp] += 1 
            else:
                # Loop through time positions from start time to end time
                for tp in np.arange(timepos1, timepos2 + 1, 1):
                    activity_time_matrix_MH[activity - 1][tp] += 1 
```
